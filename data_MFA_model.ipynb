```python
import numpy as np
from generate_data import MFA_model
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n_components = 10
dim = 3
dim_latent = 2
dim_latent_range = (1, 2)
model = MFA_model(n_components, dim, dim_latent_range=dim_latent_range, seed_rng=143)
```

```python
data, labels = model.generate_data(1000)
```

```python
%matplotlib inline
```

```python
COLORS = ['r', 'b', 'g', 'c', 'orange', 'm', 'lawngreen', 'gold', 'y', 'hotpink', 'blueviolet', 
          'crimson', 'lavender', 'khaki', 'darkseagreen', 'darkturquoise', 'brown', 'firebrick', 
          'maroon', 'orchid', 'teal', 'fuchsia', 'indigo', 'palevioletred', 'rosybrown', 'grey']
c = [COLORS[v] for v in labels]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2], marker='o', c=c, s=20, alpha=0.5)
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('x_3')
plt.show()
```
